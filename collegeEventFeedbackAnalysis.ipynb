```python
import pandas as pd

# Load the Excel file (replace filename as needed)
df = pd.read_csv('student_feedback.csv')

# Preview first few rows
df.head()
```

```python
# STEP 5: Average Ratings per Category
import matplotlib.pyplot as plt
import seaborn as sns

avg_ratings = df.mean().sort_values()

plt.figure(figsize=(10,6))
sns.barplot(x=avg_ratings.values, y=avg_ratings.index, palette='coolwarm')
plt.title('Average Student Rating per Feedback Category')
plt.xlabel('Average Rating (1 to 5)')
plt.ylabel('Feedback Category')
plt.show()
```

```python
# STEP 6: Rating Distribution Boxplot
plt.figure(figsize=(12,6))
sns.boxplot(data=df, orient='h', palette='Set2')
plt.title('Distribution of Ratings Across Feedback Categories')
plt.xlabel('Rating')
plt.show()

```

```python
# STEP 7: Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='Blues')
plt.title('Correlation Between Feedback Categories')
plt.show()
```

```python
# STEP 8: Identify Best and Worst Rated Aspects
print(" Top Rated Aspect:")
print(avg_ratings.idxmax(), "->", round(avg_ratings.max(), 2))

print("\n Lowest Rated Aspect:")
print(avg_ratings.idxmin(), "->", round(avg_ratings.min(), 2))

```

#  Summary of Insights

-  **Best Rated Aspect**: "Explains concepts in an understandable way"  students appreciate clarity.
-  **Lowest Rated Aspect**: "Degree of difficulty of assignments"  assignments might be too tough.
-  **Strong Correlation**: Found between Well versed with subject and Solves doubts willingly.
